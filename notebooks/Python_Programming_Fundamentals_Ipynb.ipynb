**Operators**

help assign, compare, and test conditions.

```python
x = 10
y = 5

print('Arithmetic Operators')
print('Addition of x and y', x + y)         # Addition. 15
print('Subtraction of x and y', x - y)      # Subtraction 5
print('Multiplication of x and y', x * y)   # Multiplication 50
print('Division of x and y', x / y)         # Division 2.0
print('Modulus of x and y', x % y)          # Modulus 0
print('x power y', x ** y, end="\n\n")      # Power 100000

print('Assignement Operators')
x += 20
print(x)  # 30
x -= 10
print(x)  # 20
x *= 2
print(x)  #40
x /= 2
print(x, end="\n\n")  #20.0

print('Comparison Operators')
print(10 == 10) #True
print(10 != 5)  #True
print(10 > 5)   #True
print(10 < 5)   #False
print(10 >= 10) #True
print(10 <= 5, end="\n\n")  #False

print('Logical Operators')
print(10 > 5 and 5 > 2)           #True
print(10 > 5 or 5 < 2)            #True
print(not (10 > 5), end="\n\n")   #False

print('Membership Operators')
print("a" in "Akanksha")       #True
print("a" not in "Akanksha")   #False
print("x" not in "Akanksha", end="\n\n")   #True

print('Identity Operators')
a = 10; b = 10
print(a is b)       #True
print(a is not b)   #False
```

**Python string functions** help modify, check, search, and format text efficiently using built-in methods.
Strings are immutable
These functions do not change the original string, they return a new one.

```python
#Case Conversion
StrFun = 'Akanksha Hello World!'
print(StrFun.upper())         #Converts to uppercase. AKANKSHA HELLO WORLD!
print(StrFun.lower())         #Converts to lowercase. akanksha hello world!
print(StrFun.capitalize())    #First letter uppercase.  Akanksha hello world!
print(StrFun.title())         #First letter of each word uppercase .  Akanksha Hello World!
print(StrFun.swapcase())      #Upper â†” Lower case.  aKANKSHA hELLO wORLD!
```

```python
#Space Removing Functions
StrFun = ' Hello Akanksha '
print(StrFun.strip())     #Removes spaces from both ends. -- Hello Akanksha
print(StrFun.rstrip())    #Removes spaces from right.     --  Hello Akanksha
print(StrFun.lstrip())    #Removes spaces from left.      -- Hello Akanksha
```

```python
#Searching & Checking Functions
StrFun = 'Hello Akanksha'
print(StrFun.find('a'))             #Finds index of substring - 8
print(StrFun.index('A'))            #Like find (error if not found) - 6
print(StrFun.startswith('hello'))   #Checks starting word - False
print(StrFun.endswith('a'))         #Checks ending word - True
print(StrFun.count('a'))            #Counts occurrences - 2
```

```python
#Type Checking functions
StrFun, StrFunU, StrFunL, StrFunD, StrFunAlNum, StrSpace= 'Akanksha', 'AKANKSHA', 'akanksha', '1234', 'ak123', ' '
print(StrFunU.isupper())        #All uppercase
print(StrFunL.islower())        #All lowercase
print(StrFun.isalpha())         #Only letters
print(StrFunD.isdigit())        #Only numbers
print(StrFunAlNum.isalnum())    #Letters + numbers
print(StrSpace.isspace())       #Only spaces
```

```python
#Replace, Split & Join
StrFun = 'I Love JAVA'

print(StrFun.replace('JAVA','Python'))    #Replace text
print(StrFun.replace('java','Python'))

print(StrFun.split(' '))                  #Split string into list

StrJoin = ['I','love','Data']

print(" ".join(StrJoin))                  #Join list into string
```
